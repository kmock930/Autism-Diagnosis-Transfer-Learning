## Read Video Data and Save them into Numpy arrays

```python
X_train_tikharm, y_train_tikharm, X_val_tikharm, y_val_tikharm, X_test_tikharm, y_test_tikharm = read_videos(
    dataset=constants.TIKHARM_DATASET
);
```

```python
print("-----TikHarm Dataset-----");
print("Training Set");
print(f"Total Number of Samples in training set: {len(X_train_tikharm)}");
print(f"Total Number of Labels in training set: {len(y_train_tikharm)}");
print(f"Dimensionality of a video sample in training set: {X_train_tikharm.ndim}");
print(f"Structure of a video sample in training set: {X_train_tikharm[random.randint(0, len(X_train_tikharm)-1)].shape}");
print(f"Unique Labels in training set: {np.unique(y_train_tikharm)}");

print("Validation Set");
print(f"Total Number of Samples in validation set: {len(X_val_tikharm)}");
print(f"Total Number of Labels in validation set: {len(y_val_tikharm)}");
print(f"Dimensionality of a video sample in validation set: {X_val_tikharm.ndim}");
print(f"Structure of a video sample in validation set: {X_val_tikharm[random.randint(0, len(X_val_tikharm)-1)].shape}");
print(f"Unique Labels in validation set: {np.unique(y_val_tikharm)}");

print("Test Set");
print(f"Total Number of Samples in test set: {len(X_test_tikharm)}");
print(f"Total Number of Labels in test set: {len(y_test_tikharm)}");
print(f"Dimensionality of a video sample in test set: {X_test_tikharm.ndim}");
print(f"Structure of a video sample in test set: {X_test_tikharm[random.randint(0, len(X_test_tikharm)-1)].shape}");
print(f"Unique Labels in test set: {np.unique(y_test_tikharm)}");
```

```python
X_train_violence, y_train_violence, X_val_violence, y_val_violence, X_test_violence, y_test_violence = read_videos(
    dataset=constants.VIOLENCE_DATASET
);
```

```python
print("-----Violence Detection Dataset-----");
print("Training Set");
print(f"Total Number of Samples in training set: {len(X_train_violence)}");
print(f"Total Number of Labels in training set: {len(y_train_violence)}");
print(f"Dimensionality of a video sample in training set: {X_train_violence.ndim}");
print(f"Structure of a video sample in training set: {X_train_violence[random.randint(0, len(X_train_violence)-1)].shape}");
print(f"Unique Labels in training set: {np.unique(y_train_violence)}");

print("Validation Set");
print(f"Total Number of Samples in validation set: {len(X_val_violence)}");
print(f"Total Number of Labels in validation set: {len(y_val_violence)}");
print(f"Dimensionality of a video sample in validation set: {X_val_violence.ndim}");
print(f"Structure of a video sample in validation set: {X_val_violence[random.randint(0, len(X_val_violence)-1)].shape}");
print(f"Unique Labels in validation set: {np.unique(y_val_violence)}");

print("Test Set");
print(f"Total Number of Samples in test set: {len(X_test_violence)}");
print(f"Total Number of Labels in test set: {len(y_test_violence)}");
print(f"Dimensionality of a video sample in test set: {X_test_violence.ndim}");
print(f"Structure of a video sample in test set: {X_test_violence[random.randint(0, len(X_test_violence)-1)].shape}");
print(f"Unique Labels in test set: {np.unique(y_test_violence)}");
```

```python
# Save the Numpy arrays into .npy files
np.save("X_train_tikharm.npy", X_train_tikharm);
np.save("y_train_tikharm.npy", y_train_tikharm);
np.save("X_val_tikharm.npy", X_val_tikharm);
np.save("y_val_tikharm.npy", y_val_tikharm);
np.save("X_test_tikharm.npy", X_test_tikharm);
np.save("y_test_tikharm.npy", y_test_tikharm);

np.save("X_train_violence.npy", X_train_violence);
np.save("y_train_violence.npy", y_train_violence);
np.save("X_val_violence.npy", X_val_violence);
np.save("y_val_violence.npy", y_val_violence);
np.save("X_test_violence.npy", X_test_violence);
np.save("y_test_violence.npy", y_test_violence);
```
